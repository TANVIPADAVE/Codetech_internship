```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn plots
sns.set(style='whitegrid')

# Ensure plots are displayed inline in notebooks
%matplotlib inline
```

```python
# Load the Titanic dataset from a URL
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
df = pd.read_csv(url)

# Display the first few rows to get an overview
df.head()
```

```python
# Get a summary of the dataset
df.info()

# Summary statistics for numerical columns
df.describe()

# Check for missing values in each column
df.isnull().sum()
```

```python
# Display columns and check for missing values
print("Columns in the DataFrame:\n", df.columns)
print("\nMissing values per column:\n", df.isnull().sum())

# Fill missing values in 'Age' with the median value of 'Age'
if 'Age' in df.columns:
    df['Age'] = df['Age'].fillna(df['Age'].median())

# Check if 'Embarked' column exists and has missing values
if 'Embarked' in df.columns:
    # Fill missing values in 'Embarked' with the mode (most common value)
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop duplicates, if any exist in the dataset
df = df.drop_duplicates()

# Verify the changes by displaying the updated info
print("\nUpdated DataFrame Information:")
df.info()
```

```python
# Count plot for 'Survived'
sns.countplot(data=df, x='Survived')
plt.title('Survival Count')
plt.show()

# Count plot for 'Pclass' (Passenger Class)
sns.countplot(data=df, x='Pclass')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Distribution of 'Age'
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of 'Fare'
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Survival rate by gender
sns.countplot(data=df, x='Survived', hue='Sex')
plt.title('Survival by Gender')
plt.show()

# Survival rate by passenger class
sns.countplot(data=df, x='Survived', hue='Pclass')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Boxplot for 'Age' by 'Survived'
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age Distribution by Survival')
plt.show()

# Boxplot for 'Fare' by 'Survived'
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Scatter plot for 'Age' vs. 'Fare'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title('Age vs. Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load your dataset (replace 'your_dataset.csv' with the actual file path)
# For example, if using a local file:
df = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")

# Step 3: Display the first few rows to confirm successful loading
print(df.head())

# Step 4: Proceed with correlation analysis
# Select only numerical columns while ignoring any completely empty columns
df_numerical = df.select_dtypes(include=[np.number])

# Drop columns that contain all NaN values, if any
df_numerical = df_numerical.dropna(axis=1, how='all')

# Check if there are at least two numerical columns to compute correlations
if df_numerical.shape[1] > 1:
    # Calculate the correlation matrix for the remaining numerical columns
    corr_matrix = df_numerical.corr()

    # Display the correlation matrix
    print("\nCorrelation Matrix:\n", corr_matrix)

    # Display the heatmap of correlations
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()
else:
    print("Not enough numerical columns for correlation analysis.")
```

```python
# List of numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Check if there are numerical columns before plotting
if len(numerical_columns) > 0:
    for col in numerical_columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data=df, x=col)
        plt.title(f'Boxplot of {col}')
        plt.show()
else:
    print("No numerical columns available for outlier detection.")
```

```python
from sklearn.preprocessing import StandardScaler

# List of numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Check if there are numerical columns before scaling
if len(numerical_columns) > 0:
    # Scale numerical features using StandardScaler
    scaler = StandardScaler()
    df_numerical_scaled = scaler.fit_transform(df[numerical_columns])

    # Convert scaled data back to a DataFrame
    df_numerical_scaled = pd.DataFrame(df_numerical_scaled, columns=numerical_columns)

    # Display the first few rows of the scaled numerical data
    print("Scaled Numerical Data (first 5 rows):")
    print(df_numerical_scaled.head())
else:
    print("No numerical columns available for scaling.")
```
